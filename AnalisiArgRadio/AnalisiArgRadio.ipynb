```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import neighbors
from sklearn.svm import SVR
from sklearn.ensemble import VotingClassifier
```

```python
df = pd.read_excel('dataset.xlsx',
                 sheet_name = 'Dataset1')
df.head()
```

```python
argomenti = []
for i in range (1, 17):
    x = "Arg"+str(i)
    argomenti.append(x)
print(argomenti)
```

```python
for i in argomenti:
    df[i] = pd.Categorical(df[i])
    
df.info()
```

```python
df.Arg1.value_counts()
```

```python
l_full_arg = {}

for i in argomenti:
     x = df[str(i)].value_counts()
     x_df = pd.DataFrame(x)
     
```

```python
l_args = []
l_pvs = []

for arg in argomenti:
    for i, txt in enumerate(df[str(arg)]):
        #print(txt, df.Pageviews[i])
        l_args.append(txt)
        l_pvs.append(df.Pageviews[i])
        
df_final = pd.DataFrame({'Argomento': l_args, 'Pageviews': l_pvs})
df_final.head(15)
```

```python
df_alt1 = pd.DataFrame(df_final.groupby('Argomento').Pageviews.sum())
df_alt1.sort_values(by=['Pageviews'],
                    inplace=True,
                    ascending=False)
df_alt1.head(15)
```
